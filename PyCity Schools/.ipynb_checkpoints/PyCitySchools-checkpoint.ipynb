```python
# Import Pandas and dependencies
import pandas as pd
```

```python
# Import CSV files
schools_file = "Resources/schools_complete.csv"
students_file = "Resources/students_complete.csv"

# Read CSV files and create DataFrame
schools_df = pd.read_csv(schools_file)
students_df = pd.read_csv(students_file)
```

```python
# Merge two DataFrames using left join
merged = pd.merge(students_df, schools_df, on=["school_name", "school_name"], how="left")
merged
```

```python
### District Summary
```

```python
# Total number of unique schools
total_schools = merged["School ID"].nunique()
total_schools
```

```python
# Total students
total_students = merged["student_name"].count()
total_students
```

```python
# Total budget - only 15 total schools
school_budget = merged["budget"].unique()
total_budget = sum(school_budget)
total_budget
```

```python
# Average math score
avg_math = merged["math_score"].mean()
avg_math
```

```python
# Average reading score
avg_reading = merged["reading_score"].mean()
avg_reading
```

```python
# % passing math (Percentage of students who passed math)
passed_math = sum(merged["math_score"] >= 70)
percent_pmath = (passed_math / total_students) * 100
percent_pmath
```

```python
# % passing reading (Percentage of students who passed reading)
passed_reading = sum(merged["reading_score"] >= 70)
percent_preading = (passed_reading / total_students) * 100
percent_preading
```

```python
# % overall passing (Percentage of students who passed math and reading)
passed_both = (merged["math_score"] >= 70) & (merged["reading_score"] >= 70)
students_passed_all = len(merged.loc[passed_both, :])
percent_all = students_passed_all / total_students * 100
percent_all 
```

```python
# Create district summary DataFrame
district_summary_values = {'total_schools':[total_schools], 'total_students':[total_students], 'total_budget':[total_budget], 'avg_math':[avg_math], 'avg_reading':[avg_reading], 'percent_pmath':[percent_pmath], 'percent_preading':[percent_preading],'percent_all':[percent_all]}
district_summary_df = pd.DataFrame(district_summary_values)
district_summary_df
```

```python
# Formatting for district summary DataFrame
# total_students - add comma
total_students_format = district_summary_df["total_students"].map('{:,.0f}'.format)

# total_buDget - add comma, $ sign, 2 decimal places
total_budget_format = "$"+district_summary_df["total_budget"].map('{:,.2f}'.format)

# Create new summary values list with proper formatting 

format_district_summary_values = {'Total Schools':total_schools, 'Total Students':total_students_format, 'Total Budget':total_budget_format, 'Average Math Score':avg_math, 'Average Reading Score':avg_reading, '% Passing Math':percent_pmath, '% Passing Reading':percent_preading, '% Overall Passing Rate':percent_all}

# Create new formatted DataFrame
district_summary_df = pd.DataFrame(format_district_summary_values)
district_summary_df
```

```python
### School Summary
```

```python
# Create a DataFrame that summarizes key metrics about each school.

# Want to create groupby as collecting data about each particular school - 'school_name'
groupedschools = merged.groupby('school_name')

# Set index to find unique values for each school
school_index = merged.set_index(["school_name"])

# Find each school name
school = merged['school_name'].unique()
school
```

```python
# School type
school_type = (groupedschools["type"].unique()).str[0]
school_type
```

```python
# Total students per school
total_students_school = (groupedschools["size"].unique()).astype('int64')
total_students_school
```

```python
# Total school budget per school
total_school_budget = (groupedschools["budget"].unique()).astype('int64')
total_school_budget
```

```python
# Per student budget for each school
per_student_budget = (total_school_budget / total_students_school).astype('float64')
per_student_budget
```

```python
# Average math score per school
avgmath_school = groupedschools["math_score"].mean()
avgmath_school
```

```python
# Average reading score per school
avgreading_school = groupedschools["reading_score"].mean()
avgreading_school
```

```python
# Percent passing math - number of students passing/total students

# Filter students with >70 
passing_math = merged['math_score'] >= 70

# Create df with that filter
students_passing_math_df = merged.loc[passing_math, :]

# New groupby for math w/ filter
grouped_math = students_passing_math_df.groupby('school_name')

# Count the groupby object to get the number of students passing
num_passing_math = grouped_math['math_score'].count()

# Percent for maths - dividing by total students
perc_math = (num_passing_math/total_students_school) * 100

# Setting number as decimal
perc_math = perc_math.astype('float64')
perc_math
```

```python
# % passing reading - number of students passing/total students

# Filter students with >70 
passing_reading = merged['reading_score'] >= 70

# Create df with that filter
students_passing_reading_df = merged.loc[passing_reading, :]


# New groupby for reading w/ filter
grouped_reading = students_passing_reading_df.groupby('school_name')

# Count the groupby object to get the number of students passing
num_passing_reading = grouped_reading['reading_score'].count()

# Percent for reading - dividing by total students
perc_reading = (num_passing_reading/total_students_school) * 100

# Setting number as decimal
perc_reading = perc_reading.astype('float64')
perc_reading 
```

```python
# % passing both math and reading = number of students(math & reading) per school/ total students per school

# Filter students >70
passing_both = (merged['math_score'] >= 70) & (merged['reading_score'] >= 70)
# Creating a df with filter
students_passing_both_df = merged.loc[passing_both, :]
# Finding the length of the df of students passing (=number of students passing)
students_passing_both = len(merged.loc[passing_both, :])


grouped_overall = students_passing_both_df.groupby('school_name')
# Count # of math/reading - columns same length, apply count to get the # of students passing 
num_passing_overall = grouped_overall['math_score'].count()
# Percent overall - dividing by total students
perc_passing_overall = num_passing_overall/total_students_school * 100

# Setting number as decimal
perc_passing_overall = perc_passing_overall.astype('float64')
perc_passing_overall
```

```python
# Create a DataFrame for School Summary
school_summary_values = {'School Type':school_type, 'Total Students per School':total_students_school, 'Total Budget per School':total_school_budget, 'per Student Budget': per_student_budget, 'Average Math Score': avgmath_school, 'Average Reading Score': avgreading_school, '% Passing Math': perc_math, '% Passing Reading': perc_reading, '% Passing Overall': perc_passing_overall}
school_summary_df = pd.DataFrame(school_summary_values)


## FORMATTING
# Formating index name
school_summary_df.index.name = None

# total_students - add comma
total_students_school_format = school_summary_df["Total Students per School"].map('{:,.0f}'.format)

# total_budget per school - add comma, $ sign, 2 decimal places
total_school_budget_format = "$"+school_summary_df["Total Budget per School"].map('{:,.2f}'.format)

# per student budget - add comma, $ sign, 2 decimal places
per_student_budget_format = "$"+school_summary_df["per Student Budget"].map('{:,.2f}'.format)


# Create new summary values list with proper formatting 
school_summary_values = {'School Type':school_type, 'Total Students per School':total_students_school_format, 'Total Budget per School':total_school_budget_format, 'per Student Budget': per_student_budget_format, 'Average Math Score': avgmath_school, 'Average Reading Score': avgreading_school, '% Passing Math': perc_math, '% Passing Reading': perc_reading, '% Passing Overall': perc_passing_overall}
# Create new DataFrame
school_summary_df = pd.DataFrame(school_summary_values)
school_summary_df
```

```python
# Highest-Performing Schools - Rank in descending order

top_schools = school_summary_df.loc[school_summary_df['% Passing Overall']>90]

# Formating index name
top_schools.index.name = None

# Create
top_schools.sort_values(['% Passing Overall'], ascending = False).head(5)
```

```python
# Lowest-Performing Schools - Rank in descending order

bottom_schools = school_summary_df.loc[school_summary_df['% Passing Overall']<90]

# Formating index name
bottom_schools.index.name = None

# Create
bottom_schools.sort_values(['% Passing Overall']).head(5)
```

```python
# Math Scores by Grade

ninth_graders = merged[(merged["grade"] == "9th")]
ninth_graders_scores = ninth_graders.groupby("school_name")['math_score'].mean()

tenth_graders = merged[(merged["grade"] == "10th")]
tenth_graders_scores = tenth_graders.groupby("school_name")['math_score'].mean()

eleventh_graders = merged[(merged["grade"] == "11th")]
eleventh_graders_scores = eleventh_graders.groupby("school_name")['math_score'].mean()

twelfth_graders = merged[(merged["grade"] == "12th")]
twelfth_graders_scores = twelfth_graders.groupby("school_name")['math_score'].mean()


#Create DataFrame
math_scores_by_grade = {'9th':ninth_graders_scores, '10th':tenth_graders_scores, '11th':eleventh_graders_scores, '12th':twelfth_graders_scores}
math_scores_by_grade_df = pd.DataFrame(math_scores_by_grade)


# Minor data wrangling
math_scores_by_grade_df.index.name = None


# Display the DataFrame
math_scores_by_grade_df
```

```python
# Reading Scores by Grade

ninth_graders = merged[(merged["grade"] == "9th")]
ninth_graders_scores = ninth_graders.groupby("school_name")['reading_score'].mean()

tenth_graders = merged[(merged["grade"] == "10th")]
tenth_graders_scores = tenth_graders.groupby("school_name")['reading_score'].mean()

eleventh_graders = merged[(merged["grade"] == "11th")]
eleventh_graders_scores = eleventh_graders.groupby("school_name")['reading_score'].mean()

twelfth_graders = merged[(merged["grade"] == "12th")]
twelfth_graders_scores = twelfth_graders.groupby("school_name")['reading_score'].mean()


#Create DataFrame
reading_scores_by_grade = {'9th':ninth_graders_scores, '10th':tenth_graders_scores, '11th':eleventh_graders_scores, '12th':twelfth_graders_scores}
reading_scores_by_grade_df = pd.DataFrame(reading_scores_by_grade)


# Minor data wrangling
reading_scores_by_grade_df.index.name = None


# Display the DataFrame
reading_scores_by_grade_df
```

```python
### Scores by School Spending
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_student_budget, bins=spending_bins, labels=labels)

# Minor data wrangling
school_spending_df.index.name = None

school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Math Score"]
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["Average Reading Score"]
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Math"]
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Reading"]
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"]).mean()["% Passing Overall"]
```

```python
# Create DataFrame
spending_summary = {'Average Math Score':spending_math_scores, 'Average Reading Score':spending_reading_scores, '% Passing Math':spending_passing_math, '% Passing Reading':spending_passing_reading, '% Overall Passing':overall_passing_spending} 
spending_summary_df = pd.DataFrame(spending_summary)

# Display results
spending_summary_df
```

```python
### Scores by School Size
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_size_df = school_summary_df.copy()
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_size_df["School Size"] = pd.cut(total_students_school, bins=size_bins, labels=labels)

# Minor data wrangling
school_size_df.index.name = None

school_size_df
```

```python
# Calculate averages for the desired columns. 
size_math_scores = school_size_df.groupby(["School Size"]).mean()["Average Math Score"]
size_reading_scores = school_size_df.groupby(["School Size"]).mean()["Average Reading Score"]
size_passing_math = school_size_df.groupby(["School Size"]).mean()["% Passing Math"]
size_passing_reading = school_size_df.groupby(["School Size"]).mean()["% Passing Reading"]
size_overall_passing = school_size_df.groupby(["School Size"]).mean()["% Passing Overall"]
```

```python
# Create DataFrame
size_summary = {'Average Math Score':size_math_scores, 'Average Reading Score':size_reading_scores, '% Passing Math':size_passing_math, '% Passing Reading':size_passing_reading, '% Overall Passing':size_overall_passing} 
size_summary_df = pd.DataFrame(size_summary)


# Minor data wrangling
size_summary_df.index.name = "School Size"

size_summary_df
```
